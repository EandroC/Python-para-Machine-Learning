```python
def maiorFrequencia():
    contador = 0
    numero = listaJogadas[0]
     
    for i in listaJogadas:
        frequencia = listaJogadas.count(i)
        if(frequencia> contador):
            contador = frequencia
            numero = i
    return numero
```

```python
def previsao():
    numero = maiorFrequencia()
    if numero == 1:
        return 2
    elif numero == 2:
        return 3
    else:
        return 1
```

```python
import random

jogador = digite()
listaJogadas = []
jogadorBot = random.randint(1, 3)
cont = 1

while jogador != 4:
    if len(listaJogadas)%5!=0 or len(listaJogadas)==0:
        jogadorBot = random.randint(1, 3)
    
    listaJogadas.append(jogador)
    
    print(cont,'ª Rodada')
    resultado(jogador, jogadorBot)
    
    jogador = digite()
    if len(listaJogadas)%5==0:
        jogadorBot = previsao()
    cont+=1
        
print('Partida Finalizada')
```
